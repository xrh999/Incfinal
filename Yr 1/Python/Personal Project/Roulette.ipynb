```python
import random

class RouletteGame:
    def __init__(self):
        self.balance = 1000  # Starting balance in fake money
        self.numbers = list(range(37))  # Numbers on the roulette wheel (0 to 36)
        self.red_numbers = {1, 3, 5, 7, 9, 12, 14, 16, 18, 21, 23, 25, 27, 30, 32, 34, 36}
    
    def spin_wheel(self):
        return random.choice(self.numbers)
    
    def display_balance(self):
        print(f"Your current balance is: ${self.balance}")
    
    def place_bet(self):
        bet_amount = float(input("Enter your bet amount: $"))
        if bet_amount <= 0:
            print("Bet amount must be greater than zero.")
            return 0
        if bet_amount > self.balance:
            print("Insufficient balance.")
            return 0
        return bet_amount
    
    def bet_on_number(self, bet_amount):
        chosen_number = int(input("Choose a number to bet on (0-36): "))
        if chosen_number not in self.numbers:
            print("Invalid number.")
            return 0
        winning_number = self.spin_wheel()
        print(f"The roulette wheel landed on: {winning_number}")
        if chosen_number == winning_number:
            print("Congratulations! You've won!")
            return bet_amount * 35  # Roulette pays 35 to 1 on a number bet
        else:
            print("Sorry, you lost.")
            return -bet_amount
    
    def bet_on_color(self, bet_amount):
        color = input("Bet on color (red/black): ").strip().lower()
        if color not in ["red", "black"]:
            print("Invalid color choice.")
            return 0
        winning_number = self.spin_wheel()
        print(f"The roulette wheel landed on: {winning_number}")
        if (color == "red" and winning_number in self.red_numbers) or \
           (color == "black" and winning_number not in self.red_numbers and winning_number != 0):
            print("Congratulations! You've won!")
            return bet_amount
        else:
            print("Sorry, you lost.")
            return -bet_amount

    def play_game(self):
        while True:
            self.display_balance()
            bet_type = input("Choose bet type (number/color/quit): ").strip().lower()
            if bet_type == "quit":
                print("Thank you for playing!")
                break
            elif bet_type == "number":
                bet_amount = self.place_bet()
                if bet_amount > 0:
                    self.balance += self.bet_on_number(bet_amount)
            elif bet_type == "color":
                bet_amount = self.place_bet()
                if bet_amount > 0:
                    self.balance += self.bet_on_color(bet_amount)
            else:
                print("Invalid bet type.")

if __name__ == "__main__":
    game = RouletteGame()
    game.play_game()
```
